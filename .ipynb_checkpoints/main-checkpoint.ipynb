```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_healthy = pd.read_csv("./base.csv")
```

```python
df_healthy
# df means Dataframe a data structure built into pandas.
# Dataframes are two dimensional arrays, displaying data in tabular form.
# Dataframes, I think, will be extensively used for data processing
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html?highlight=dataframe#pandas.DataFrame
```

```python
df_healthy.plot()
# All 10k values plotted not much useful here
```

```python
df_healthy.iloc[0:100].plot()
# First 100 values plotted to get some more information
```

```python
df_healthy.iloc[0:100].plot(x='time',y='ax')
df_healthy.iloc[0:100].plot(x='time',y='az')
df_healthy.iloc[0:100].plot(x='time',y='ay')
# Individual axes plotted with time on x-axis and actual data value on y-axis 
```

```python
# Taking 100 points from each axis along with time
df_healthy_x_100_points = df_healthy[['time','ax']].iloc[0:200]
df_healthy_y_100_points = df_healthy[['time','ay']].iloc[0:200]
df_healthy_z_100_points = df_healthy[['time','az']].iloc[0:200]
```

```python
# making a genric function to get ffts and fft freq for each axis
# fft freq is 100 because only 100 points are taken
def get_fft_and_fftfreq(df_axis, df_time):
    fft_healthy_100 = np.fft.fft(a=df_axis)
    fft_healthy_100_freq = np.fft.fftfreq(df_time)
    return fft_healthy_100, fft_healthy_100_freq
```

```python
fft_figure = plt.figure()
fft_figure.set_size_inches(15,5)

# FFT plotted for x axis
fft_value, fft_freq_value = get_fft_and_fftfreq(df_healthy_x_100_points['ax'], 200)
subplot_x = fft_figure.add_subplot(131)
subplot_x.plot(fft_freq_value,fft_value.real,fft_freq_value,fft_value.imag)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('FFT for x axis')

# FFT plotted for y axis
fft_value, fft_freq_value = get_fft_and_fftfreq(df_healthy_y_100_points['ay'], 200)
subplot_y = fft_figure.add_subplot(132)
subplot_y.plot(fft_freq_value,fft_value.real,fft_freq_value,fft_value.imag)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('FFT for y axis')

# FFT plotted for z axis
fft_value, fft_freq_value = get_fft_and_fftfreq(df_healthy_z_100_points['az'], 200)
subplot_z = fft_figure.add_subplot(133)
subplot_z.plot(fft_freq_value,fft_value.real,fft_freq_value,fft_value.imag)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('FFT for z axis')

plt.savefig('healthy.pdf')
plt.show()
```
